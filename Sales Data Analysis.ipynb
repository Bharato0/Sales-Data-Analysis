## Spliting the Order Date column into year(contain only year) and month(contain only month) for better understanding and Visualization

# Analyze sales trends over time.

```python
# Group by year and month and sum the sales
data_grouped = data.groupby(['year', 'month'])['Sales'].sum().reset_index()

# Plot the data using a line plot
plt.figure(figsize=(12, 6))

# Iterate through each year and plot the monthly sales
for year in data_grouped['year'].unique():
    data_year = data_grouped[data_grouped['year'] == year]
    plt.plot(data_year['month'], data_year['Sales'], marker='o', label=str(year))

plt.title('Monthly Sales Trend by Year')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(range(1, 13))  # Assuming months are 1 to 12
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Based on the graph, we can observe that sales are relatively low at the beginning of the year. However, they steadily increase over the subsequent months, reaching their peak during November and December. There is a notable downturn in sales in July. Additionally, sales have shown a consistent upward trend over the years, with noticeable growth from 2011 through 2014.

### Grouping data by product name and aggregate sales

```python
product_sales = data.groupby('Product Name')['Sales'].sum().reset_index()
product_sales
```

### Sorting products by total sales

```python
product_sales_sorted = product_sales.sort_values(by='Sales', ascending=False)
product_sales_sorted
```

# Visualizing the best-selling products (top 10 for example)

```python
top_products = product_sales_sorted.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_products['Product Name'], top_products['Sales'], color='skyblue')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.title('Top 10 Best-Selling Products')
plt.gca().invert_yaxis()  # Invert y-axis to display the best sellig product on top
plt.show()
```

## From the above graph, we can see the top-selling product among them is the Apple Smart Phone, Full Size. Additionally, we can observe that the top 3 best-selling products are all phones.
